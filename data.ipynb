```python
allData = []
for i in data1:
    allData.append(i.find_all('td'))
col = []
r = [i.text for i in allData[0]][3:14]
for i in range(len(allData[1])):
    try:
        allData[1][i+3] = r[i]
    except:
        pass
for i in allData[1]:
    try:
        col.append(i.text)
    except:
        col.append(i)
print(col)
```
